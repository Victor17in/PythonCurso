```
#B= v/c = beta
#y= 1/(1-(v/c)**2)**(1/2) = 1/(1-(B)**2)**(1/2)
import matplotlib.pyplot as plt

def gamma(beta):
  #global y # sem isso não há print()
  y = 1/((1-(float(beta))**2)**(1/2))
  return y

gamma(0.6)
#print("Valor de gamma:", y)

#beta vai de 0 a 0.995, fazer grafico yxbeta
beta = 0
betalimit = 0.995
betalist = []
gammalist = []
gammabetalist = []

while beta <= betalimit:
  betalist.append(beta)
  gamma(beta)
  gammalist.append(gamma(beta))
  gammabetalist.append(beta*(gamma(beta)))
  beta = beta + 0.005

print(gammalist)
print(betalist)
print(gammabetalist)

#grafico gamma x beta
dadosdegamma = gammalist
dadosdebeta=betalist
plt.plot(dadosdegamma,dadosdebeta)
plt.xlabel("gamma")
plt.ylabel("beta")

#grafico gamma*beta x beta SEPARAR DO OUTRO

dadosdegammabeta = gammabetalist
dadosdebeta=betalist
plt.plot(dadosdegammabeta,dadosdebeta)
plt.xlabel("gamma*beta")
plt.ylabel("beta")
```
