````
#5
%%writefile lerarquivoscolunas.py
def ler_arquivos(file):
  try:
    fin = open(file)
  except Exception as err:
    print(type(err))
    print(err)
    print('Arquivo não encontrado')
    return 
  fin = open(file)
  linhas = fin.readlines()
  #print (linhas)
  global listaidade;
  global listaaltura;
  global listapeso;
  global listacounter;
  listaidade = []
  listaaltura = []
  listapeso = []
  listacounter = []
  counter = 0
  for line in linhas:
    #print (line)
    column = line.strip().split('\t')
    #print (column)
    counter += 1
    listaidade.append(float(column[0]))
    listaaltura.append(float(column[1]))
    listapeso.append(float(column[2]))
    listacounter.append(float(counter))
    
  print('Coluna 1: ',listaidade)
  print('Coluna 2: ',listaaltura)
  print('Coluna 3: ',listapeso)
  return listaidade, listaaltura, listapeso, listacounter
ler_arquivos('dados_alunos.txt')
````

````
#7
import lerarquivoscolunas
import matplotlib.pyplot as plt
import statistics as stat
from lerarquivoscolunas import listaidade
from lerarquivoscolunas import listaaltura
from lerarquivoscolunas import listapeso
from lerarquivoscolunas import listacounter
idade= listaidade
altura= listaaltura
peso= listapeso
counter = listacounter

midade = stat.mean(listaidade)
maltura = stat.mean(listaaltura)
mpeso = stat.mean(listapeso)
print('Média da idade:', midade, 'anos')
print('Média da altura:', maltura, 'm')
print('Média do peso(massa):',mpeso, 'kg')
dpopuidade = stat.pstdev(listaidade)
dpopualtura = stat.pstdev(listaaltura)
dpopupeso = stat.pstdev(listapeso)
print('Desvio populacional da idade:',dpopuidade, 'anos')
print('Desvio populacional da altura:',dpopualtura, 'm')
print('Desvio populacional do peso:',dpopupeso, 'kg')
damoidade = stat.stdev(listaidade)
damoaltura = stat.stdev(listaaltura)
damopeso = stat.stdev(listapeso)
print('Desvio amostral da idade:',damoidade, 'anos')
print('Desvio amostral da altura:',damoaltura, 'm')
print('Desvio amostral do peso:',damopeso, 'kg')
print('Idade =', round(midade,2), '+-', round(damoidade,2), 'anos')
print('Altura =', round(maltura,2), '+-', round(damoaltura,2), 'm')
print('Peso(massa) =', round(mpeso,2), '+-', round(damopeso,1), 'kg')
````
